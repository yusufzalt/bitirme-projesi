```python
import tkinter as tk
from tkinter import messagebox
import cv2
import numpy as np
import os
from PIL import Image,ImageTk



window = tk.Tk()
window.title('Face Recognizer')
window.geometry('800x200')
window.resizable(False,False)

l1 = tk.Label(window,text = 'Ad     :',font = ('Algerian ',20))
l1.grid(column = 0, row = 0)
t1 = tk.Entry(window,width = 50 ,bd = 5)
t1.grid(column = 1, row = 0)

l2 = tk.Label(window,text = 'yas    :',font = ('Algerian ',20))
l2.grid(column = 0, row = 1)
t2 = tk.Entry(window,width = 50 ,bd = 5)
t2.grid(column = 1, row = 1)

l3 = tk.Label(window,text = 'Adres :',font = ('Algerian ',20))
l3.grid(column = 0, row = 2)
t3 = tk.Entry(window,width = 50 ,bd = 5)
t3.grid(column = 1, row = 2)

def verTopla():
    face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)
    if(t1.get() == '' or t2.get() == '' or t3.get() ==''):
        messagebox.showinfo('messag','Lutfen Ad, Yas ve Adres Kisimlari doldurunuz!!!')
    else :
        canvas=tk.Canvas(self.window,width=900,height=450)
        canvas.pack()
        veri_topla_form = tk.Toplevel()
        veri_topla_form.title('Veri Toplama Formu')
        veri_topla_form.geometry('700x400')
        veri_topla_form.resizable(False,False)

        frame1 = tk.Frame(veri_topla_form,highlightbackground='gray',highlightcolor='gray',highlightthickness=1.5,bd=10)
        frame1.place(x=0,y=0, width=450,height=250)

        label1 = tk.Label(frame1,width=430,height=230)
        label1.place(x=0, y=0)
        def videoOku(cap=0):
            ret,frame = cap.read()
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_classifier.detectMultiScale(gray,1.3,5)
            for (x,y,w,h) in faces:
                cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2) 

            from_array = Image.fromarray(img)
            photo = ImageTk.PhotoImage(image=from_array)
            label1.configure(image=photo)
            label1.image=photo
            label.after(20,vedeoOku)
            
        videoOku(cap)

    
    
    def kirp():
        
        
        def face_cropped(img):
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_classifier.detectMultiScale(gray,1.3,5)

            # scaling(yukselme) = 1.3
            # Minimum Neighbor(en yakin komsu) = 5

            if faces is():
                return None
            for(x,y,w,h) in faces:
                cropped_face = img[y:y+h, x:x+w]
            return cropped_face
        
        
        id = 1
        img_id = 0

        while True :
            ret, frame = cap.read()
            if face_cropped(frame) is not None:
                img_id += 1
                face = cv2.resize(face_cropped(frame),(200,200))
                face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

                file_name_path = 'data/user.'+str(id)+'.'+str(img_id)+'.jpg'
                cv2.imwrite(file_name_path,face)
                
                
                #cv2.putText(face,str(img_id),(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,255),2)

                #cv2.imshow('Cropped Face',face)
                if cv2.waitKey(1) == 27 or int(img_id) == 50:
                    break
        
        cap.release()
        cv2.destroyAllWindows()
        messagebox.showinfo('bidiri','Veri Toplama Islemi Tamamlandi ...')
        t1.delete(0, 'end')
        t2.delete(0, 'end')
        t3.delete(0, 'end')
    

b1 = tk.Button(window, text = 'Veri Topla', font = ('Algerian',20), bg = 'green',fg = 'white',command=verTopla)
b1.grid(column = 0, row = 4)

def train():
    data_path = 'C:/Users/abdul/Desktop/Face_Detected/data'
    path = [os.path.join(data_path,f) for f in os.listdir(data_path)]
    faces = []
    ids = []
    
    for image in path:
        img = Image.open(image).convert('L')
        imageNp=np.array(img,'uint8')
        id = int(os.path.split(image)[1].split('.')[1])
        
        faces.append(imageNp)
        ids.append(id)
    ids = np.array(ids)
    
    # Train the Classifier and save
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.train(faces,ids)
    clf.write('classifier.xml')
    messagebox.showinfo('massag','Egitim Tamamlandi ...')



b2 = tk.Button(window, text = 'Egit', font = ('Algerian',20), bg = 'yellow',fg = 'black',command = train)
b2.grid(column = 1, row = 4)

def face_detect():
    def draw_boundary(img,classifier, scaleFactor, minNeighbors, color, text, clf):
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        features = classifier.detectMultiScale(gray_img,scaleFactor,minNeighbors)

        coords = []

        for(x,y,w,h) in features:
            cv2.rectangle(img,(x,y),(x+w, y+h), color, 2)

            id, pred = clf.predict(gray_img[y:y+h,x:x+w])
            confidences = int(100 * (1-pred/300))

            if confidences > 70:
                if id == 1:
                    cv2.putText(img,'settar',(x,y-5),cv2.FONT_HERSHEY_SIMPLEX,0.8,color,1,cv2.LINE_AA)
            else:
                cv2.putText(img,'Bilinmeyen',(x,y-5),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,0,255),1,cv2.LINE_AA)

            coords = [x,y,w,h]
        return coords

    def recogniz(img,clf,faceCascade):
        coords = draw_boundary(img, faceCascade, 1.1, 10, (255,255,255), 'face', clf )
        return img

    face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.read('classifier.xml')

    Video_capture = cv2.VideoCapture(0)

    while True:
        ret, frame = Video_capture.read()
        frame = recogniz(frame,clf,face_classifier)

        cv2.imshow('Faces',frame)

        key=cv2.waitKey(1)
        if key == 27 :
            break

    Video_capture.release()
    cv2.destroyAllWindows()

b3 = tk.Button(window, text = 'Yuz Tanima', font = ('Algerian',20), bg = 'blue',fg = 'white',command = face_detect)
b3.grid(column = 2, row = 4)

def cikis():
    window.destroy()

b4 = tk.Button(window, text = 'Cikis', font = ('Algerian',20), bg = 'red',fg = 'white',command = cikis)
b4.grid(column = 4, row = 4)

b4 = tk.Button(window, text = '', font = ('Algerian',20))
b4.grid(column = 3, row = 4)

window.mainloop()
```
